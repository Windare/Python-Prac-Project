```python
#check how well the orders are delivered if there is any problem with delivery.
sb.set(rc = {'figure.figsize':(10,4)})
fig1 = sb.countplot(x = 'order_status', data = KPMG_data_sheet1)
fig1 = fig1.get_figure()   
fig1.savefig('orderstatus.png')
```

```python
#to check for most sort after brand and the product size by custormers
sb.set(rc = {'figure.figsize':(15,8)})
fig2 = sb.countplot(x = 'brand', hue = "product_size", data = KPMG_data_sheet1)
fig2 = fig2.get_figure()   
fig2.savefig('MOst sort after brand.png', dpi=100)
```
